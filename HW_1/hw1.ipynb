## Ερώτημα 2: Υπολογισμός παραγώγων

###2.1

Έχουμε $f(x) = x^TAx + b^Tx$ με $A=A^T$ διάστασης $dxd$ και $b\in R^d$ \\
(1) $\frac{	\partial b^Tx}{	\partial x} = \frac{	\partial x^Tb}{	\partial x} = b$ \\
(2) $\frac{	\partial x^TAx}{	\partial x} = (A+A^T)x = 2Ax$ επειδή $A=A^T$ \\
Από (1) + (2) έχουμε $\nabla f(x) = 2Ax + b$

###2.2

Το πρόβλημα βελτιστοποίησης που δίνεται είναι: $min_x||A-XB||_F^2$ \\
Για να υπολογίσουμε το ολικό ελαχιστό θα υπολογίσουμε το: $\nabla x ||A-XB||_F^2$ \\
 $||A-XB||_F^2 = tr((A-XB)^T(A-XB)) = tr((A^T-X^TB^T)(A-XB) = \\
  A^TA-A^TXB-B^TX^TA+B^TX^TXB = A^TA - 2B^TX^TA + B^TX^TXB$ \\
Xρησιμοποιούμε την ιδιότητα της συμμετρίας του πίνακα X^TX για να αντικαταστήσουμε τον όρο B^TX^TXB με B^T(X^TX)B άρα προκύπτει: \\
$Α^TΑ - 2Β^TX^TΑ + Β^T(X^TX)Β$ \\
Υπολογίζουμε την παράγωγο του: \\
1. $\nabla x (Α^TΑ) = 0$ \\
2. $\nabla x (-2Β^TX^TΑ) = -2Β^ΤA$ \\
3. $\nabla xΒ^T(X^TX)Β) = 2Β^T(X^TX)$ \\

Από 1,2,3 έχουμε $\nabla x= -2Β^ΤA +2Β^T(X^TX)$ \\
Λύνουμε $\nabla x=0$ ως προς και X: \\
$-2Β^ΤA +2Β^T(X^TX)=0 ⇒ \\
2Β^T(X^TX) = 2Β^TΑ ⇒ \\
X^TX = (Β^T)^{-1}Β^TΑ ⇒ \\
X = (Β^TΒ)^{-1}Β^TΑ$ \\


## Ερώτημα 3: Gradient descent

```python
import matplotlib.pyplot as plt
import numpy as np


def gradient_descent(curX1, curX2, lrate, maxIter, f, dfX1, dfX2, funName):
  precision = 0.0001
  result = []
  print("Gradient Descent for", funName, "with starting values", (curX1, curX2), "learning rate", lrate, "and max iter", maxIter)
  for i in range(maxIter):
    prevX1 = curX1
    prevX2 = curX2
    curX1 = curX1 - (lrate * dfX1(curX1,curX2))
    curX2 = curX2 - (lrate * dfX2(curX1,curX2))
    result.append(f(curX1, curX2))
    print("Iter:", i+1," f(x1,x2) =",f(curX1, curX2))
    if (abs(curX1-prevX1) <= precision and abs(curX2-prevX2) <= precision):
      print("The local minimum occurs at", (curX1, curX2), " after", i+1, " iterations with value", f(curX1, curX2))
      break

  #create plot
  plt.title("rate = %.3f maxIters = %d"%(lrate, maxIter))
  plt.xlabel("iterations")
  plt.ylabel("f1(x1,x2)")
  plt.plot([x for x in range(0, i+1)], result)
  plt.show()
```

###Συνάρτηση $f_1 = (x_1-2)^2 + (x_2-3)^2$

```python
def f1(x1,x2):
  return ((x1 - 2)**2 + (x2 - 3)**2)

def df1x1(x1,x2):
  return (2*x1 - 4)

def df1x2(x1,x2):
  return (2*x2 - 6)
```

Για f1 με αρχικές τιμές (0,0) με learning rate 0.5 και μέγιστο αριθμό επαναλήψεων 10 παρατηρούμε ότι βρίσκει κατευθείαν το τοπικό ελάχιστο το οποίο είναι το 0

```python
gradient_descent(0,0, 0.5, 10, f1, df1x1, df1x2, "f1")
```

Αρχικά αυξάνουμε τον αριθμό μέγιστών επαναλήψεων και τον κρατάμε σταθερό. Παρατηρούμε ότι όσο απομακρινόμαστε από το ιδανικό
 learning rate = 0.5 είτε αυξανοντάς το είτε μειώνοντας το αυξάνεται ο αριθμός επαναλήψεων όπως φαίνεται παρακάτω: \\
learning rate = 0.01  iterations = 318 \\
learning rate = 0.1  iterations = 40 \\
learning rate = 0.25  iterations = 15 \\
learning rate = 0.5  iterations = 1 \\
learning rate = 0.65 iterations = 10 \\
learning rate = 0.8 iterations = 23 \\
learning rate = 0.99 iterations = 546 \\


```python
gradient_descent(0,0, 0.01, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(0,0, 0.1, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(0,0, 0.25, 1000, f1, df1x1, df1x2, "f1")
```

```python
gradient_descent(0,0, 0.65, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(0,0, 0.8, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(0,0, 0.99, 1000, f1, df1x1, df1x2, "f1")
```

Εάν αλλάξουμε τις αρχικές τιμές $x_1, x_2$ παρατηρούμε ότι: \\
Εάν το learning rate το έχουμε σταθερά στο 0.5 (την ιδανική τιμή του) παρατηρούμε ότι δεν έχει μεγάλη σημασία οι αρχικές τιμές και βρίσκει το ελάχιστο
σε μικρό αριθμό επαναλήψεων αυτό συμβαίνει γιατί η συνάρτηση είναι σχετικά απλή και έχει επιλεχτεί το ιδανικό learning rate. \\
Εάν δεν χρησιμοποιούμε ιδανικό learning rate πχ το 0.25 και το κρατάμε σταθερό παρατηρούμε ότι όσο απομακρίνομαστε από τις τιμές του ολικού ελαχίστου αυξάνεται ο αριθμός των επαναλήψεων


```python
gradient_descent(30,20, 0.5, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(-500,-200, 0.5, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(0,0, 0.5, 1000, f1, df1x1, df1x2, "f1")
```

```python
gradient_descent(300, 200, 0.25, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(30,20, 0.25, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(-500,-200, 0.25, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(-20,-100, 0.25, 1000, f1, df1x1, df1x2, "f1")
gradient_descent(0,0, 0.25, 1000, f1, df1x1, df1x2, "f1")
```

###Συνάρτηση $f_2 = (1 - (x_2-3))^2 + 20((x_1 + 3) - (x_2 - 3)^2)^2$

```python
def f2(x1,x2):
  return ((1- (x2 - 3))**2 + 20*((x1 + 3) - (x2 - 3)**2)**2)

def df2x1(x1,x2):
  return (40*(x1 + 3 - (x2 - 3)**2))

def df2x2(x1,x2):
  return (-2*(4-x2) + 40*(x1 + 3 -(x2-3)**2)*(-2*x2 + 6))
```

Για f2 με αρχικές τιμές (0,0) με learning rate 0.5 και μέγιστο αριθμό επαναλήψεων 10 παρατηρούμε ότι συμβαίνει overflow και για να το αποφύγουμε πρέπει να μικρίνουμε το learning rate

```python
gradient_descent(0,0, 0.5, 100, f2, df2x1, df2x2, "f2")
```

Παρατηρούμε ότι το learning rate πρέπει να είναι πολύ μικρό για να αποφυγούμε το overflow η ιδανική τιμή του είναι κοντά στο 0.005 και όσο μικραίνουμε τον αριθμό αυξάνονται οι επαναλήψεις πιο συγκεκριμένα: \\
learning rate = 0.005 iterations = 1511 \\
learning rate = 0.001 iterations = 4937 \\
learning rate = 0.0005 iterations = 7607 \\


```python
gradient_descent(0,0, 0.005, 10000, f2, df2x1, df2x2, "f2")
gradient_descent(0,0, 0.001, 10000, f2, df2x1, df2x2, "f2")
gradient_descent(0,0, 0.0005, 10000, f2, df2x1, df2x2, "f2")
```

Η συνάρτηση $f_2$ για learning rate=0.5 κοντά στο ιδανικό παρουσιάζει μεγαλύτερες διακυμάνσεις με την αλλάγη των αρχικών τιμών επείδη η συνάρτηση f2 είναι πιο ευαίσθητη σε αλλάγες. Όμοια με την $f_1$ όσο απομακρινόμαστε από τις τιμές ολικού ελαχίστου αυξάνεται ο αριθμός των επαναλήψεων.

```python
gradient_descent(0,0, 0.005, 10000, f2, df2x1, df2x2, "f2")
gradient_descent(-3, 3, 0.005, 10000, f2, df2x1, df2x2, "f2")
gradient_descent(-1, 5, 0.005, 10000, f2, df2x1, df2x2, "f2")
```
